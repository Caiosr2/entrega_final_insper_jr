```python
import pandas as pd
import matplotlib.pyplot as plt


anos = [2013, 2014, 2015, 2016]

for ano in anos:
    df_ano = pd.read_excel('dados_amr.xlsx', sheet_name=f'Pedidos_{ano}')
    contagem_subcategorias = df_ano['subcategoria'].value_counts()

    plt.pie(contagem_subcategorias, labels=contagem_subcategorias.index, wedgeprops=dict(width=0.5))
    plt.title(f'Distribuição de Subcategorias - Pedidos {ano}')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

anos = [2013, 2014, 2015, 2016]

for ano in anos:
    df_ano = pd.read_excel('dados_amr.xlsx', sheet_name=f'Pedidos_{ano}')
    contagem_categorias = df_ano['categoria'].value_counts()

    plt.pie(contagem_categorias, labels=contagem_categorias.index, wedgeprops=dict(width=0.5), autopct='%1.1f%%')
    plt.title(f'Distribuição de Categorias - Pedidos {ano}')
    plt.show()
```
