```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir o caminho do arquivo e os anos
file_path = "dados_amr.xlsx"
anos = [2013, 2014, 2015, 2016]

# Unir todos os anos em um único DataFrame
df_list = [pd.read_excel(file_path, sheet_name=f'Pedidos_{ano}') for ano in anos]
df_all = pd.concat(df_list)

# Agrupar por região e segmento e somar as quantidades
resumo = df_all.groupby(['região', 'segmento'])['quantidade'].sum().unstack()

# Plotar gráfico de barras agrupadas
resumo.plot(kind='bar', figsize=(10, 6))
plt.title('Vendas por Segmento em Cada Região (2013-2016)')
plt.xlabel('Região')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.legend(title='Segmento')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados combinados de todos os anos
file_path = "dados_amr.xlsx"
anos = [2013, 2014, 2015, 2016]

df_list = [pd.read_excel(file_path, sheet_name=f'Pedidos_{ano}') for ano in anos]
df_all = pd.concat(df_list)

# Obter as regiões únicas
regioes = df_all['região'].unique()

# Gerar um gráfico separado para cada região
for regiao in regioes:
    df_regiao = df_all[df_all['região'] == regiao]
    
    # Agrupar por subcategoria e somar as quantidades
    resumo = df_regiao.groupby('subcategoria')['quantidade'].sum().sort_values(ascending=False)
    
    # Plotar gráfico de barras
    plt.figure(figsize=(8, 6))
    resumo.plot(kind='bar')
    plt.title(f'Produtos Mais Vendidos na Região {regiao} (2013-2016)')
    plt.xlabel('Produto (Subcategoria)')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=90)
    plt.grid(axis='y')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar todos os anos juntos
file_path = "dados_amr.xlsx"
anos = [2013, 2014, 2015, 2016]

df_list = [pd.read_excel(file_path, sheet_name=f'Pedidos_{ano}') for ano in anos]
df_all = pd.concat(df_list)

# Obter todas as regiões únicas
regioes = df_all['região'].unique()

# Gerar gráfico por região
for regiao in regioes:
    df_regiao = df_all[df_all['região'] == regiao]
    
    # Contar quantas vezes cada valor de desconto foi aplicado
    contagem_descontos = df_regiao['desconto'].value_counts().sort_index()
    
    # Plotar gráfico de barras
    plt.figure(figsize=(8, 6))
    contagem_descontos.plot(kind='bar')
    plt.title(f'Frequência de Descontos Aplicados - Região {regiao} (2013-2016)')
    plt.xlabel('Valor do Desconto')
    plt.ylabel('Quantidade de Pedidos')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```
