```python
import pandas as pd
import matplotlib.pyplot as plt

anos = [2013, 2014, 2015, 2016]

for ano in anos:
    df_ano = pd.read_excel('dados_amr.xlsx', sheet_name=f'Pedidos_{ano}')
    
    # Contar quantidade de pedidos por estado
    contagem_estados = df_ano['estado'].value_counts().sort_values(ascending=False)
    
    # Gerar o gráfico de barras
    plt.figure(figsize=(10, 5))
    contagem_estados.plot(kind='bar')
    plt.title(f'Quantidade de Pedidos por Estado - {ano}')
    plt.xlabel('Estado')
    plt.ylabel('Quantidade de Pedidos')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
